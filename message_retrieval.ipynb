```python
convo_id = 'XXX'
limit = 500
until = '2014-01-01'
base_url = 'https://graph.facebook.com/%s/' % (convo_id)
url = '%s?limit=%s&access_token=%s' % (base_url,limit,ACCESS_TOKEN)
content = requests.get(url).json()
print(json.dumps(content,indent=1))
```

```python
# retrieving and saving msgs

msgs = content
file_name = 'output.csv'
outputFile = open(file_name,'w',newline='')
outputWriter = csv.writer(outputFile)

# headers
outputWriter.writerow(['convo_id','timestamp',
                      'sender_id','sender_name',
                     'msg_id','msg_contents'])

participants = msgs['to']['data']
convo_id = msgs['id']

# intitialize first set of msgs
current_messages = msgs['comments']
counter = 0
num_messages = 0
num_people = 1

# loop through all message pages and write them to a file
while True:
    counter += 1
    num_messages += len(current_messages['data'])
    try:
        next_page = current_messages['paging']['next']
        print("page %d with %d messages read" % (counter, num_messages))
    except:
        break
    for item in current_messages['data']:
        created_time = item['created_time']
        sender_id = item['from']['id']
        sender_name = item['from']['name']
        msg_id = item['id']
        try:
            contents = item['message'] # fix encoding problem
        except: # if no msg or unknown character
            contents = '<NO_MESSAGE_FOUND>'
        row = [convo_id,created_time,sender_id,sender_name,msg_id,contents.encode('utf-8')]
        outputWriter.writerow(row)
    # change limit on next_page
    next_page = next_page.replace('?limit=25','?limit=500')
    #update current_messages    
    current_messages = requests.get(next_page).json()
outputFile.close()
```
